```python
# Now choosing the fixed features that might affect house prices

house1 = house[['MSZoning','LotFrontage', 'LotArea', 'Street','LotShape', 'LandContour', 'Utilities', 'LotConfig',
                'LandSlope','Neighborhood','Condition1', 'Condition2', 'BldgType', 'HouseStyle','YearRemodAdd',
                'TotalBsmtSF','1stFlrSF', '2ndFlrSF','GrLivArea','BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
                'BedroomAbvGr', 'KitchenAbvGr','TotRmsAbvGrd','GarageType','GarageCars','GarageArea','YrSold','SalePrice'
                ]].copy()

house1.head()
```

```python
# I choose 30 features from my intuition and some reserach on real state economy 

house1.shape
```

```python

# plot heatmap to check correlation

data_corr = house1.corr()
# Set the default matplotlib figure size:
fig, ax = plt.subplots(figsize=(15, 10))
# Generate a mask for the upper triangle (taken from seaborn example gallery)
mask = np.zeros_like(data_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# Plot the heatmap with seaborn.
# Assign the matplotlib axis the function returns. This will let us resize the labels.
ax = sns.heatmap(data_corr, mask=mask, ax=ax, cmap='Blues')
# Resize the labels.
ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=14)
ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=14)
plt.show()
# sns.heatmap(residential.corr(),cmap='Blues')


```

```python
# heatnmap correlation values :

data_corr['SalePrice'].sort_values(ascending=False).head(10)
# => Get the first 10 variables that have higher correlation
```

```python
# Preparing the dataset for train set split
# the training data will contain the pre 2010 data
# the test data will contain the 2010 data

print(house1['YrSold'].unique())

train_data = house1[house1['YrSold'] != 2010]

print(train_data['YrSold'].unique())


# getting the test data


test_data = house1[house1['YrSold'] == 2010]

print(test_data['YrSold'].unique())
```

```python
# getting the dummy variables for the categorical data


train_data = pd.get_dummies(train_data,
                columns=['MSZoning', 'Street','LotShape', 'LandContour',
                        'Utilities', 'LotConfig','LandSlope','Neighborhood',
                         'Condition1', 'Condition2', 'BldgType', 'HouseStyle',
                         'YearRemodAdd','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','GarageType',
                        'GarageCars','YrSold'], drop_first=True)
                                            
train_data.head()
```

```python
# now that I have my cleaned train data pre2010, lets define our target and features

# I need to get the index of the target variable which is our sale price

print(train_data.columns.get_loc("SalePrice"))

y = train_data.iloc[:,11].values

print(y)

X = train_data.iloc[:, train_data.columns != 'SalePrice']

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

# Use the "fit_transform" function to standardize the X design matrix.
Xs = ss.fit_transform(X)

```

# regression algorithms

```python
# applying linear regression

linreg = LinearRegression()

linreg_scores = cross_val_score(linreg, Xs, y, cv=10)

print(linreg_scores)
print(np.mean(linreg_scores))


# the r^2 is negative which is meaning less
```

```python
# applying regularization with ridge 

# finding an optimal value for the ridge regression alpha

ridge_alphas = np.logspace(-2, 7, 50)

optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(Xs, y)

print(optimal_ridge.alpha_)
```

```python
# cross validate the ridge regression

ridge = Ridge(alpha=optimal_ridge.alpha_)

ridge_scores = cross_val_score(ridge, Xs, y, cv=10)

print(ridge_scores)
print(np.mean(ridge_scores))
```

```python
# It's vastly better than the linear regression. 
# There's likely so much multicollinearity in the data that linear regression gives meaningless coefficients
# on predictors.
#  Ridge is able to manage the multicollinearity and get a good out-of-sample result.
```

```python
#  optimal value for lasso regression alpha 

optimal_lasso = LassoCV(n_alphas=500, cv=10, verbose=1)
optimal_lasso.fit(Xs, y)

print(optimal_lasso.alpha_)

```

```python
lasso = Lasso(alpha=optimal_lasso.alpha_)

lasso_scores = cross_val_score(lasso, Xs, y, cv=10)

print(lasso_scores)
print(np.mean(lasso_scores))
```

```python

# The lasso performs almost same to ridge, lasso - .78 and ridge = .78

# lets look at the weights of the predictors in Lasso
```

```python
lasso.fit(Xs, y)
```

```python
lasso_coefs = pd.DataFrame({'variable':X.columns,
                            'coef':lasso.coef_,
                            'abs_coef':np.abs(lasso.coef_)})

lasso_coefs.sort_values('abs_coef', inplace=True, ascending=False)

lasso_coefs.head(20)
```

```python
#  Compare the residuals for ridge and lasso visually.
ridge.fit(Xs, y)
lasso.fit(Xs, y)

ridge_resid = y - ridge.predict(Xs)
lasso_resid = y - lasso.predict(Xs)

sns.jointplot(ridge_resid, lasso_resid)

```

```python
# now we want to check our model on the test data

#  first lets fetch the test data which is already prepared


test_data = pd.get_dummies(test_data,
                columns=['MSZoning', 'Street','LotShape', 'LandContour',
                        'Utilities', 'LotConfig','LandSlope','Neighborhood',
                         'Condition1', 'Condition2', 'BldgType', 'HouseStyle',
                         'YearRemodAdd','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','GarageType',
                        'GarageCars','YrSold'], drop_first=True)
                                            
test_data.head()
```

```python
# checking the model performance on test data : preparing the test data

print(test_data.columns.get_loc("SalePrice"))

y_test = test_data.iloc[:,11].values

X_test = test_data.iloc[:, test_data.columns != 'SalePrice']


from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

# Use the "fit_transform" function to standardize the X design matrix.

X_tests = ss.fit_transform(X_test)

lasso.fit(X_tests, y_test)

predicted_y = lasso.predict(X_tests)

print(predicted_y)
```

```python
# getting the scores on test data

lasso = Lasso(alpha=optimal_lasso.alpha_)

lasso_scores = cross_val_score(lasso, X_tests, y_test, cv=10)

print(lasso_scores)
print(np.mean(lasso_scores))
```

```python

residual_data = pd.DataFrame(list(zip(y_test, predicted_y)),
                                 columns=['Actual SalePrice', 'Predicted SalePrice'])

sns.lmplot(x='Actual SalePrice', y='Predicted SalePrice', data=residual_data)
plt.title('Prediction of SalePrice with Fixed features only')
plt.show()
```
